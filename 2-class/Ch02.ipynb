You are an engineering manager with 30 engineers working for you.  18 are located in Pittsburgh and 12 are located in Chicago.  You need to select 4 to attend the annual conference.  To be fair, you draw names from a hat.  What is the probability that you pick 2 employees from Pittsburgh and 2 employees from Chicago?


```python
pittsburgh = 18
chicago = 12
pick_total = 4

favorable = comb(pittsburgh, 2) * comb(chicago, 2)
total = comb(pittsburgh + chicago, pick_total)

prob = favorable / total
prob
```

Additive Rule #3 (Complements)
What is the probability that a student is 5’5” or taller?

```python
# Frequencies by height interval
freq = {
    "5'0''–5'1''": 2,
    "5'2''–5'4''": 6,
    "5'5''–5'7''": 7,
    "5'8''–5'10''": 14,
    "5'11''–6'1''": 10,
    "6'2''–6'4''": 6,
    ">6'4''": 0
}

# Total number of students
total = sum(freq.values())

# Complement: students shorter than 5'5''
shorter_than_5_5 = freq["5'0''–5'1''"] + freq["5'2''–5'4''"]

# Probability using complement rule
p_at_least_5_5 = 1 - shorter_than_5_5 / total

print(f"P(height ≥ 5'5'') = {p_at_least_5_5:.3f}")
```

In how many ways could the three homes be chosen if the order of visiting is considered?

In how many ways could the three homes be chosen if the order is disregarded?

1 in 1000 adults have a rare disease. There is a test for the disease. Let's say the test is 99\% accurate. If a randomly selected individual tests positive, what is the probability that he or she actually has the disease?

```python
P_D = 0.001
P_not_D = 1- P_D
P_T_given_D = 0.99
P_T_given_not_D = 1- P_T_given_D

P_T = P_T_given_D * P_D + P_T_given_not_D * P_not_D

P_D_given_T = (P_T_given_D * P_D) / P_T

print(P_T)
print(P_D_given_T)
```

```python
DT = 0.99 * 0.001        # P(T and D)
D_notT = 0.01 * 0.001    # P(T' and D)
notD_T = 0.01 * 0.999    # P(T and D')
notD_notT = 0.99 * 0.999 # P(T' and D')

# Totals
total_T = DT + notD_T
total_notT = D_notT + notD_notT

# Conditional probability
P_D_given_T = DT / total_T

print("Joint probability table:")
print(f"P(T and D)   = {DT:.5f}")
print(f"P(T and D')  = {notD_T:.5f}")
print(f"P(T)       = {total_T:.5f}")
print(f"P(D | T) = {P_D_given_T:.4f}")
```

On a given night, LeBron James plays against one of three types of opponents:
$A_1$: weak defensive team, $P(A_1)=0.4$
$A_2$: average defensive team, $P(A_2)=0.35$
$A_3$: elite defensive team, $P(A_3)=0.25$. Let $B$ be the event that LeBron scores 30 or more points. Historical data:
$
P(B\mid A_1)=0.6,\quad P(B\mid A_2)=0.45,\quad P(B\mid A_3)=0.25$.

What is (a) $P(A_2 \cap B)$?  (b) $P(B)$ on a random night?  (c) If LeBron scored 30+ points, what type of game was most likely?


```python
# Probabilities of opponent type
P_A1 = 0.40   # weak defense
P_A2 = 0.35   # average defense
P_A3 = 0.25   # elite defense

# Conditional probabilities used in the solution slides
P_B_given_A1 = 0.6
P_B_given_A2 = 0.45
P_B_given_A3 = 0.25

P_A2_and_B = P_B_given_A2 * P_A2
print(f"P(A2 ∩ B) = {P_A2_and_B:.3f}")

P_B = (
    P_B_given_A1 * P_A1 +
    P_B_given_A2 * P_A2 +
    P_B_given_A3 * P_A3
)
print(f"P(B) = {P_B:.3f}")

P_A1_given_B = (P_B_given_A1 * P_A1) / P_B
P_A2_given_B = (P_B_given_A2 * P_A2) / P_B
P_A3_given_B = (P_B_given_A3 * P_A3) / P_B

print(f"P(A1 | B) = {P_A1_given_B:.3f}")
print(f"P(A2 | B) = {P_A2_given_B:.3f}")
print(f"P(A3 | B) = {P_A3_given_B:.3f}")
```
