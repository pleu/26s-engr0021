## Die Experiment
A 6-sided die is rolled 100 times and the sum is 365. Can this be explained by chance variation alone, or does it
suggest that the die is loaded?

```python
import math
from statistics import mean
import numpy as np
from scipy import stats
from math import sqrt, ceil

n = 100
observed_sum = 365

# Fair die
expected_single = sum(range(1, 7)) / 6
expected_sum = n * expected_single

var_single = sum((x - expected_single)**2 for x in range(1, 7)) / 6
sd_single = math.sqrt(var_single)
sd_sum = math.sqrt(n * var_single)

z = (observed_sum - expected_sum) / sd_sum
p_two_sided = 2 * (1 - stats.norm.cdf(abs(z)))

p_two_sided
```

## ESP experiment
A machine with 4 targets is guessed 7500 times, with 2006 correct guesses. Can this be explained by chance variation alone, or does it suggest that something else is happening here.

```python
n = 7500
p0 = 1/4
observed = 2006

expected = n * p0
sd = math.sqrt(n * p0 * (1 - p0))
z = (observed - expected) / sd
p_upper = 1 - stats.norm.cdf(z)

p_upper
```

## Is the coin fair?
A coin is tossed 100 times and 61 heads are observed. Test
\(H_0: theta = 0.5\) vs. \(H_1: theta > 0.5\).

```python
n = 100
x = 61
theta0 = 0.5

expected = n * theta0
se = math.sqrt(n * p0 * (1 - theta0))
z = (x - expected) / se
p_upper = 1 - stats.norm.cdf(z)

p_upper
```

## Light bulb lifetime
- $\bar{x}$ = 738.44
- s = 38.2
- n = 10
- Test \(H_0: \mu = 750\) vs \(H_1: \mu < 750\)
- $\alpha = 0.05$


```python
xbar = 738.44
mu0 = 750
s = 38.2
n = 10
alpha = 0.05

t_stat = (xbar - mu0) / (s / sqrt(n))
df = n - 1
p = stats.t.cdf(t_stat, df)   # left-tailed
t_crit = stats.t.ppf(alpha, df)

t_stat, t_crit, p
```

## GPA and car ownership

- Non-owners: $n_1=100, \bar{x}_1=2.70, \sigma_1^2=0.36$
- Car owners: $n_2=100, \bar{x}_2=2.54, \sigma_2^2=0.40$
- Test $H_0: \mu_1-\mu_2=0\) vs \(H_1: \mu_1-\mu_2 \neq 0$
- $\alpha = 0.05$


```python
n1, xbar1, sigma1_sq = 100, 2.70, 0.36
n2, xbar2, sigma2_sq = 100, 2.54, 0.40
d0 = 0
alpha = 0.05

se = sqrt(sigma1_sq / n1 + sigma2_sq / n2)
z_stat = ((xbar1 - xbar2) - d0) / se
p = 2 * (1 - stats.norm.cdf(abs(z_stat)))
z_crit = stats.norm.ppf(1 - alpha / 2)

z_stat, z_crit, p
```

## Ultrasonic humidifiers
Brand 1 outputs:
- 14.0, 14.3, 12.2, 15.1

Brand 2 outputs:
- 12.1, 13.6, 11.9, 11.2

Test $H_0: \mu_1-\mu_2=0$ vs $H_1: \mu_1-\mu_2 \neq 0$, assuming equal variances.


```python
brand1 = np.array([14.0, 14.3, 12.2, 15.1])
brand2 = np.array([12.1, 13.6, 11.9, 11.2])

n1, n2 = len(brand1), len(brand2)
xbar1, xbar2 = brand1.mean(), brand2.mean()
s1, s2 = brand1.std(ddof=1), brand2.std(ddof=1)

sp2 = ((n1 - 1)*s1**2 + (n2 - 1)*s2**2) / (n1 + n2 - 2)
sp = sqrt(sp2)
df = n1 + n2 - 2
se = sp * sqrt(1/n1 + 1/n2)

t_stat = (xbar1 - xbar2) / se
p = 2 * (1 - stats.t.cdf(abs(t_stat), df))
t_crit = stats.t.ppf(1 - 0.05/2, df)

t_stat, t_crit, p

```

## Fabric porosity

- Cotton: $n_1=10, \bar{x}_1=51.71, s_1=0.79$
- Triacetate: $n_2=10, \bar{x}_2=136.14, s_2=3.59$


```python
n_cotton, xbar_cotton, s_cotton = 10, 51.71, 0.79
n_tri, xbar_tri, s_tri = 10, 136.14, 3.59
diff = xbar_cotton - xbar_tri
se = sqrt(s_cotton**2/n_cotton + s_tri**2/n_tri)
df = (s_cotton**2/n_cotton + s_tri**2/n_tri)**2 / (
    (s_cotton**2/n_cotton)**2/(n_cotton - 1) +
    (s_tri**2/n_tri)**2/(n_tri - 1)
)
t_stat = diff / se
p_two_sided = 2 * (1 - stats.t.cdf(abs(t_stat), df))
p_for_tri_greater = stats.t.cdf(t_stat, df)  # because cotton - tri < 0 iff tri > cotton

t_crit = stats.t.ppf(1 - 0.05/2, df)

t_stat, t_crit, p_for_tri_greater
```

## Fabric breaking load

Data from the notes:

Unabraded:
- 36.4, 55.0, 51.5, 38.7, 43.2, 48.8, 25.6, 49.8

Abraded:
- 28.5, 20.0, 46.0, 34.5, 36.5, 52.5, 26.5, 46.5

Use differences = unabraded - abraded and test:
- $H_0: \mu_D = 0$
- $H_1: \mu_D > 0$

```python
_unabraded = np.array([36.4, 55.0, 51.5, 39.7, 43.2, 48.8, 25.6, 49.8])
abraded   = np.array([28.5, 20.0, 46.0, 34.5, 36.5, 52.5, 26.5, 46.5])
alpha = 0.01
d = unabraded - abraded
n = len(d)
dbar = d.mean()
sd = d.std(ddof=1)
se = sd / sqrt(n)
t_stat = dbar / se
df = n - 1
p = 1 - stats.t.cdf(t_stat, df)  # right-tailed
t_crit = stats.t.ppf(1 - alpha/2, df)

t_stat, t_crit, p
```

## Rat weight gain variability

- Control: \(n=23, s=32\)
- Low-dose: \(n=20, s=54\)

Test:
- H_0: \sigma_{low}^2 = \sigma_{control}^2
- H_1: \sigma_{low}^2 > \sigma_{control}^2
- $\alpha = 0.05$


```python
n_low, s_low = 20, 54
n_control, s_control = 23, 32
alpha = 0.05

F = (s_low**2) / (s_control**2)
df1 = n_low - 1
df2 = n_control - 1
p = 1 - stats.f.cdf(F, df1, df2)
F_crit = stats.f.ppf(1 - alpha, df1, df2)

F, F_crit, p
```

## Loaded die example — chi-square goodness-of-fit test

- Observed counts: `[4, 6, 17, 16, 8, 9]`
- Total throws: `60`
- Null hypothesis: the die is fair, so each outcome has probability `1/6`


```python
observed = np.array([4, 6, 17, 16, 8, 9])
n = observed.sum()
k = len(observed)

expected_probs = np.repeat(1/6, 6)
expected = n * expected_probs

chi2_stat = ((observed - expected)**2 / expected).sum()
df = k - 1  # no parameters estimated
p_value = 1 - stats.chi2.cdf(chi2_stat, df)
critical_value = stats.chi2.ppf(0.95, df)
```
