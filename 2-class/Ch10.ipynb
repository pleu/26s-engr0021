## Die Experiment
A 6-sided die is rolled 100 times and the sum is 365. Can this be explained by chance variation alone, or does it
suggest that the die is loaded?

```python
import math
from statistics import mean
from scipy import stats

n = 100
observed_sum = 365

# Fair die
expected_single = sum(range(1, 7)) / 6
expected_sum = n * expected_single

var_single = sum((x - expected_single)**2 for x in range(1, 7)) / 6
sd_single = math.sqrt(var_single)
sd_sum = math.sqrt(n * var_single)

z = (observed_sum - expected_sum) / sd_sum
p_two_sided = 2 * (1 - stats.norm.cdf(abs(z)))

p_two_sided
```

## ESP experiment
A machine with 4 targets is guessed 7500 times, with 2006 correct guesses. Can this be explained by chance variation alone, or does it suggest that something else is happening here.

```python
n = 7500
p0 = 1/4
observed = 2006

expected = n * p0
sd = math.sqrt(n * p0 * (1 - p0))
z = (observed - expected) / sd
p_upper = 1 - stats.norm.cdf(z)

p_upper
```

## Is the coin fair?
A coin is tossed 100 times and 61 heads are observed. Test
\(H_0: p = 0.5\) vs. \(H_1: p > 0.5\).

```python
n = 100
x = 61
p0 = 0.5

expected = n * p0
se = math.sqrt(n * p0 * (1 - p0))
z = (x - expected) / se
p_upper = 1 - stats.norm.cdf(z)

p_upper
```
