## Exponential Distribution ##

**Bridge Example**

Stress ranges in certain bridge connections can be modeled using an exponential
distribution with mean value 6 MPa. What is the probability that the stress range is at
most 10 MPa?

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Given mean (beta parameter)
beta = 6  # MPa
# Value of interest
x = 10  # MPa

x = np.linspace(0, 30, 500)
y = (1 / beta) * np.exp(-x / beta)

p_le_10 = 1 - np.exp(-10 / beta)
print(p_le_10)

# Plot PDF
plt.figure()
plt.plot(x, y, label="PDF")

```

Use the standard normal table to determine:

If $X$ is normal with $\mu=5$ and $\sigma^2=4$, determine $P(1<X<8)$.

```python
mu = 5
sigma2 = 4
sigma = math.sqrt(sigma2)

# Bounds
x1 = 1
x2 = 8

# Convert to z-scores
z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

# Compute probability
p = norm.cdf(z2) - norm.cdf(z1)
print(p)

# Plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Normal PDF")
```

Given a normal distribution with $\mu=20$ and $\sigma^2=81$, find $x$ that has:
* 19\% of the area to the left
* 32\% of the area to the right

Use norm.ppf which is percent point function

```python
from scipy.stats import norm
import math

mu = 20
sigma2 = 81
sigma = math.sqrt(sigma2)

p_left_1 = 0.19          # 19% to the left
p_right_2 = 0.32         # 32% to the right
p_left_2 = 1 - p_right_2 # convert to left area

# Find z-scores
z1 = norm.ppf(p_left_1)
z2 = norm.ppf(p_left_2)

# Convert to x-values
x1 = mu + sigma * z1
x2 = mu + sigma * z2

print(x1, x2)

# Plot
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = norm.pdf(x, mu, sigma)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Normal PDF")
```

The average time to assemble a car in a certain plant is normal with mean $20$ hours and standard deviation $2$ hours.
What is the probability that a car can be assembled in:
* less than 19.5 hours?
* between 20 and 22 hours?


```python
mu = 20      # mean (hours)
sigma = 2    # standard deviation (hours)

p1 = norm.cdf(19.5, mu, sigma)
p2 = norm.cdf(22, mu, sigma) - norm.cdf(20, mu, sigma)

print(p1, p2)

# Create x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = norm.pdf(x, mu, sigma)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label="Normal PDF")
```
