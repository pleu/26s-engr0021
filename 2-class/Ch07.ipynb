## Problem: Sum of Normal Random Variables

The time that it takes a randomly selected rat of a certain subspecies to find its way through a maze is a normally distributed random variable with mean $\mu = 1.5 \text{ minutes}$ \
and standard deviation
$\sigma = 0.35 \text{ minutes}$.

Suppose five rats are selected. Let \($X_1, X_2, \ldots, X_5$\) denote their times in the maze. Assume the \(X_i\)'s form a random sample from this normal distribution.

Define the total time for the five rats as

$Y = X_1 + X_2 + X_3 + X_4 + X_5$.

What is the probability that the total time for the five rats is between 6 and 8 minutes?

```python
import math
from scipy.stats import norm

mu = 1.5
sigma = 0.35
n = 5

mu_Y = n * mu
sigma_Y = math.sqrt(n * sigma**2)


lower = 6
upper = 8

probability = norm.cdf(upper, mu_Y, sigma_Y) - norm.cdf(lower, mu_Y, sigma_Y)

print("Mean of Y:", mu_Y)
print("Std dev of Y:", sigma_Y)
print("Probability:", probability)
```
