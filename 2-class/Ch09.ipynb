1) Review: SD of sum vs average (coin tosses)
From the slide: for a Bernoulli(p) with **p = 0.5**, the SD of a single toss is
$\sigma = \sqrt{pq} = 0.5$.

- SD of the **sum** of n tosses: $\sigma\sqrt{n}$
- SD of the **average** of n tosses: $\sigma/\sqrt{n}$

```python
import numpy as np
from math import sqrt
import math

p = 0.5
q = 1 - p
sigma = sqrt(p*q)
for n in [100, 10_000]:
    sd_avg = sigma / sqrt(n)
    sd_sum = sigma * sqrt(n)
    print(f"n={n:>5}: sigma={sigma:.3f}, SD(average)={sd_avg:.3f}, SD(sum)={sd_sum:.1f}")
```

## Keyboard height (90% CI, $\sigma$ known)
Given: $n=31$, $\bar x=80.0$ cm, $\sigma=2.0$ cm, 90% CI.

```python
from scipy.stats import norm

n = 31
x_bar = 80.0
sigma = 2.0
alpha = 0.10
z = norm.ppf(1 - alpha/2)

# Standard error
SE = sigma / sqrt(n)

# Margin of error
margin = z * SE

# Confidence interval
lower = x_bar - margin
upper = x_bar + margin

lower, upper
```

## Synthetic diamond weights (95% CI, $\sigma$ unknown)
Data: 0.46, 0.61, 0.52, 0.48, 0.57, 0.54. Assume normal.

Use the $t$ interval: $\bar x \pm t_{\alpha/2, n-1}\, s/\sqrt{n}$.

```python
from scipy.stats import t

diamonds = [0.46,0.61,0.52,0.48,0.57,0.54]
n = len(diamonds)
x = np.array(diamonds)
x_bar = x.mean()

s= x.std(ddof=1) # must use degree of freedom of 1 for sample standard deviation

sum_sq = np.sum((x - x_bar)**2)
s2 = math.sqrt(sum_sq / (n - 1))

alpha = 0.05
df = n - 1

t_crit = t.ppf(1 - alpha/2, df)

# Standard error
SE = s / math.sqrt(n)

# Margin of error
margin = t_crit * SE

# Confidence interval
lower = x_bar - margin
upper = x_bar + margin

lower, upper
```

## Hot dogs: 95% CI for mean vs 95% PI for next observation
Data (% fat): 25.2, 21.3, 22.8, 17.0, 29.8, 21.0, 25.5, 16.0, 20.9, 19.5.

- CI half-width uses $t\,s/\sqrt{n}$
- PI half-width uses $t\,s\sqrt{1+1/n}$

```python
# Data (% fat)
data = [25.2, 21.3, 22.8, 17.0, 29.8, 21.0, 25.5, 16.0, 20.9, 19.5]

x = np.array(data)
n = len(x)

# Sample mean
x_bar = x.mean()

# Sample standard deviation (use n-1)
s = x.std(ddof=1)

# Confidence level
alpha = 0.05
df = n - 1

# t critical value
t_crit = t.ppf(1 - alpha/2, df)

# ---- Confidence Interval for mean ----
SE_mean = s / math.sqrt(n)
CI_half = t_crit * SE_mean

CI_lower = x_bar - CI_half
CI_upper = x_bar + CI_half

# ---- Prediction Interval for next observation ----
PI_half = t_crit * s * math.sqrt(1 + 1/n)

PI_lower = x_bar - PI_half
PI_upper = x_bar + PI_half

CI_lower, CI_upper, PI_lower, PI_upper
```

## Car ownership and GPA (95% CI for $\mu_1-\mu_2$, variances known)
Given: $\sigma_1^2=0.36$, $n_1=100$, $\bar x_1=2.70$; $\sigma_2^2=0.40$, $n_2=100$, $\bar x_2=2.54$.

```python
sigma1_sq = 0.36
sigma2_sq = 0.40

n1 = 100
n2 = 100

xbar1 = 2.70
xbar2 = 2.54

sigma1 = math.sqrt(sigma1_sq)
sigma2 = math.sqrt(sigma2_sq)

diff = xbar1 - xbar2

SE = math.sqrt(sigma1_sq/n1 + sigma2_sq/n2)

alpha = 0.05

z = norm.ppf(1 - alpha/2)

margin = z * SE

lower = diff - margin
upper = diff + margin

lower, upper
```

## Ultrasonic humidifiers (pooled-$t$ 95% CI)
Brand 1: 14.0, 14.3, 12.2, 15.1

Brand 2: 12.1, 13.6, 11.9, 11.2

Assumptions: independent samples, normal populations, **equal variances**.

```python
brand1 = [14.0, 14.3, 12.2, 15.1]
brand2 = [12.1, 13.6, 11.9, 11.2]

x1 = np.array(brand1)
x2 = np.array(brand2)

n1 = len(x1)
n2 = len(x2)

mean1 = x1.mean()
mean2 = x2.mean()

s1 = x1.std(ddof=1)
s2 = x2.std(ddof=1)

sp2 = ((n1 - 1)*s1**2 + (n2 - 1)*s2**2) / (n1 + n2 - 2)
sp = math.sqrt(sp2)

SE = sp * math.sqrt(1/n1 + 1/n2)

df = n1 + n2 - 2

alpha = 0.05

t_crit = t.ppf(1 - alpha/2, df)

diff = mean1 - mean2

margin = t_crit * SE

lower = diff - margin
upper = diff + margin

lower, upper
```

## Fabric air permeability (Welch 95% CI from summary stats)
Cotton: $n=10$, mean=51.71, sd=0.79
Tracetate: $n=10$, mean=136.14, sd=3.59

Compute a 95% CI for $\mu_1-\mu_2$ using Welch's method.

```python
n1 = 10
mean1 = 51.71
sd1 = 0.79

n2 = 10
mean2 = 136.14
sd2 = 3.59

diff = mean1 - mean2

SE = math.sqrt((sd1**2)/n1 + (sd2**2)/n2)

num = ((sd1**2)/n1 + (sd2**2)/n2)**2
den = ((sd1**2)/n1)**2/(n1-1) + ((sd2**2)/n2)**2/(n2-1)
df = num / den

alpha = 0.05

t_crit = t.ppf(1 - alpha/2, df)

margin = t_crit * SE

lower = diff - margin
upper = diff + margin

lower, upper
```

## Paired observations: Slide vs Digital retrieval (paired-$t$ 95% CI)
Differences (Slide - Digital): [5, 19, 25, 10, 10, 10, 28, 46, 25, 38, 14, 23, 14]

Compute $\bar d$, $s_d$, and the 95% CI: $\bar d \pm t\, s_d/\sqrt{n}$.

```python
d = [5, 19, 25, 10, 10, 10, 28, 46, 25, 38, 14, 23, 14]

x = np.array(d)
n = len(x)

# Mean of differences
d_bar = x.mean()

# Sample standard deviation of differences (use n-1)
s_d = x.std(ddof=1)

# Confidence level
alpha = 0.05
df = n - 1

# t critical value
t_crit = t.ppf(1 - alpha/2, df)

# Standard error
SE = s_d / math.sqrt(n)

# Margin of error
margin = t_crit * SE

# Confidence interval
lower = d_bar - margin
upper = d_bar + margin

d_bar, s_d, lower, upper
```

## CI for $\sigma^2$ (chi-square): Concrete strength variance (90% CI)
Given: $n=10$, sample variance $s^2=195$. Find a 90% CI for $\sigma^2$.

Using slide convention with right-tail chi-square critical values.

```python
from scipy.stats import chi2

n = 10
s2 = 195
alpha = 0.10

df = n - 1

chi2_right_upper = chi2.ppf(1 - alpha/2, df)   # χ²_(α/2) right tail
chi2_right_lower = chi2.ppf(alpha/2, df)       # χ²_(1-α/2) right tail

lower = (df * s2) / chi2_right_upper
upper = (df * s2) / chi2_right_lower

lower, upper
```

## Ratio of variances ($F$ CI): Crack repair polymers (90% CI)
Epoxy: 1.75, 2.12, 2.05, 1.97
MMA prepolymer: 1.77, 1.59, 1.70, 1.69

Compute a 90% CI for $\sigma^2_{\mathrm{Epoxy}} / \sigma^2_{\mathrm{MMA}}$.

```python
from scipy.stats import f

epoxy = [1.75, 2.12, 2.05, 1.97]
mma = [1.77, 1.59, 1.70, 1.69]

x1 = np.array(epoxy)
x2 = np.array(mma)

n1 = len(x1)
n2 = len(x2)

s1_sq = x1.var(ddof=1)
s2_sq = x2.var(ddof=1)

ratio = s1_sq / s2_sq

alpha = 0.10

df1 = n1 - 1
df2 = n2 - 1

F_lower = f.ppf(alpha/2, df1, df2)
F_upper = f.ppf(1 - alpha/2, df1, df2)

lower = ratio / F_upper
upper = ratio / F_lower

lower, upper
```
