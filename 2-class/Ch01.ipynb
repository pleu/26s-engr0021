**Example - Mean**

Grade point averages of 15 University of Pittsburgh Engineering Freshmen:Grade point averages of 15 University of Pittsburgh Engineering Freshmen:


```python
import statistics

gpas = [
    2.0, 2.7, 2.8, 2.3, 3.1,
    2.8, 2.9, 2.6, 1.9, 2.7,
    3.0, 3.1, 2.6, 2.5, 2.7
]
average_gpa = sum(gpas) / len(gpas)
average_gpa
```

**Example - Median**

```python
median_gpa = statistics.median(gpas)
print(median_gpa)

numbers = [10, 14, 15, 16, 16, 16, 17, 24, 27, 29, 32, 34, 39, 47]
median_numbers = statistics.median(numbers)
print(median_numbers)
```

**Example - Median**

```python
mode_gpa = statistics.mode(gpas)
mode_gpa
```

**Mean and Median â€“ Sensitivity to Outliers**

```python
gpa_new = gpas.copy() # must use copy to make sure they are not identical
gpa_new[11] = 33.1
average_gpa = sum(gpa_new) / len(gpa_new)
print(average_gpa)

median_gpa = statistics.median(gpa_new)
median_gpa
```

**Standard Deviation and Variance**

**Example 1**

```python
observations = [4, 6, 9, 13, 18]
mean = statistics.mean(observations)
data_range = max(observations) - min(observations)
print(data_range)

variance = statistics.variance(observations)
print(variance)
standard_deviation = statistics.stdev(observations)
standard_deviation
```

**Example 2**

```python
observations = [4, 5, 9, 17, 18]
variance = statistics.variance(observations)
print(variance)
standard_deviation = statistics.stdev(observations)
standard_deviation
```

**Stem and Leaf Display**

```python
gpas_sorted = sorted(gpas)

stems = {}
for gpa in gpas_sorted:
    stem = int(gpa)
    leaf = int(round((gpa - stem) * 10))
    stems.setdefault(stem, []).append(leaf)

print("Stem | Leaves")
print("--------------")
for stem in sorted(stems):
    leaves = " ".join(str(leaf) for leaf in stems[stem])
    print(f" {stem}   | {leaves}")
```

**Histogram and Frequency Distribution**

```python

import matplotlib.pyplot as plt
import numpy as np

gpas = [
    2.0, 3.0, 2.5, 3.4, 2.1,
    2.3, 2.5, 1.4, 1.9, 2.1,
    2.9, 2.8, 2.8, 2.2, 2.3,
    2.7, 2.8, 2.6, 2.3, 2.2,
    2.6, 1.9, 3.0, 2.4, 2.1,
    2.7, 3.1, 2.7, 2.5, 2.7,
    3.1, 2.7, 3.8, 2.6, 2.8,
    2.6, 3.5, 3.0, 2.7, 2.9
]

bins = np.arange(1.0, 4.1, 1.0)
plt.figure()
plt.hist(gpas, bins=bins)
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.title("Histogram of GPAs")
plt.show()
```

```python
plt.hist(gpas, bins=bins, density=True)
plt.xlabel("GPA")
plt.ylabel("Relative Frequency")
plt.title("Relative Frequency Histogram of GPAs")
plt.show()
```
